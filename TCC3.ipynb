```python
df = pd.read_csv('BENFORD 3-3.csv', sep=';', thousands='.', decimal=',')
df.head()
```

```python
df.dropna(how='all', inplace=True)
df.head()
```

```python
y_12 = bf.Analysis(df['2012'], dec=0)
```

```python
f1d_12 = y_12.first_digits(1, conf_level=99,show_high_Z='all', ret=True)
```

```python
def plot_d(df, x, y_Exp, y_Found, N, figsize, conf_Z, text_x=False):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    # plt.title('')
    # plt.xlabel('')
    # plt.ylabel(' (%)')
    ax.bar(x, y_Found * 100., label='Encontrada')
    ax.plot(x, y_Exp * 100., color='g',linewidth=2.5, label='Benford')
    ax.legend()
    if text_x:
        plt.xticks(x,df.index, rotation='vertical')
    # Plotting the Upper and Lower bounds considering the Z for the
    # informed confidence level
    sig = conf_Z * np.sqrt(y_Exp*(1-y_Exp)/N) 
    upper = y_Exp + sig + (1/(2*N))
    lower = y_Exp - sig - (1/(2*N))
    ax.plot(x, upper * 100., color= 'r')
    ax.plot(x, lower * 100., color= 'r')
    ax.fill_between(x, upper * 100.,lower * 100., color='r', alpha=.3)
    plt.tick_params(labelsize=20)
    plt.show()
```

```python
plot_d(f1d_12, x = f1d_12.index, y_Exp = f1d_12.Expected, y_Found = f1d_12.Found,\
			 N = len(y_12), figsize = (10,8), conf_Z = 2.576)
```

```python
sd_12 = y_12.second_digit(conf_level=99,show_high_Z='all', plot=False, ret=True)
```

```python
plot_d(sd_12, x = sd_12.index, y_Exp = sd_12.Expected, y_Found = sd_12.Found,\
			 N = len(y_12), figsize = (10,8), conf_Z = 2.576)
```

```python
f2d_12 = y_12.first_digits(2, conf_level=99,show_high_Z='all', plot=False, ret=True)
```

```python
plot_d(f2d_12, x = f2d_12.index, y_Exp = f2d_12.Expected, y_Found = f2d_12.Found,\
			 N = len(y_12), figsize = (20,12), conf_Z = 2.576)
```
